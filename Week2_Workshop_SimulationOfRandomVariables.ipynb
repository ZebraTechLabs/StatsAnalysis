```
import numpy as np
import matplotlib.pyplot as plt
```

```
#For Normal Observations
from scipy.stats import norm
xNorm = norm.ppf(np.arange(1,100)/100)  #equivalent for qnorm in R - gives percentile for a probability- percent point function
yNorm = norm.cdf(xNorm) #equivalent of pnorm in R - gives probability for a number
plt.figure( figsize=(10,10) )
plt.scatter(xNorm,yNorm)
plt.hlines(yNorm,xNorm.min(),xNorm.max(),linestyle='solid',linewidth = 0.5)
plt.vlines(xNorm,yNorm.min(),yNorm.max(),linestyle='solid',linewidth = 0.5)

plt.axhline(yNorm[30],c='cyan',linewidth = 3.0)
plt.axvline(xNorm[30],c='cyan',linewidth=3.0)
```

```
#For Exponential Observations
from scipy.stats import expon
xExp = expon.ppf(np.arange(1,100)/100)  #equivalent for qnorm in R - gives percentile for a probability- percent point function
yExp = expon.cdf(xExp) #equivalent of pnorm in R - gives probability for a number
plt.figure( figsize=(10,10) )
plt.scatter(xExp,yExp)
plt.hlines(yExp,xExp.min(),xExp.max(),linestyle='solid',linewidth = 0.5)
plt.vlines(xExp,yExp.min(),yExp.max(),linestyle='solid',linewidth = 0.5)

plt.axhline(yExp[60],c='cyan',linewidth = 3.0)
plt.axvline(xExp[60],c='cyan',linewidth=3.0)
```
