```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Linear Model
#Y = aX +b +eps, a=1,b=2.5
```

```
a = 1
b = 2.5
```

```
nSample = 500
```

```
np.random.seed(0)
```

```
rand = np.random.RandomState(42)
eps = rand.normal(loc=0,scale=1.5,size=nSample)
print(eps[1:10])
```

```
#Model1
rand = np.random.RandomState(42)
X1 = rand.normal(loc=3,scale=2.5,size=nSample)
Y1 = a*X1 + b + eps
model1 = pd.DataFrame(data={'X1':X1,'Y1':Y1,'Eps':eps})
plt.scatter(model1['X1'],model1['Y1'])
```

```
#Model1 with different seed
rand = np.random.RandomState(0)
X1 = rand.normal(loc=3,scale=2.5,size=nSample)
Y1 = a*X1 + b + eps
model1 = pd.DataFrame(data={'X1':X1,'Y1':Y1,'Eps':eps})
plt.scatter(model1['X1'],model1['Y1'])
```

```
print(model1.head())
print(model1.shape)
```

```
#Model2 - with exponential distribution
rand = np.random.RandomState(0)
greekLambda = 0.5
X2 = rand.exponential(scale=1/greekLambda,size=nSample)
Y2 = a*X2 + b + eps
model2 = pd.DataFrame(data={'X2':X2,'Y2':Y2,'Eps':eps})
plt.scatter(model2['X2'],model2['Y2'])
```

```
#Model2 - with uniform distribution
rand = np.random.RandomState(0)
X2 = rand.rand(nSample)
Y2 = a*X2 + b + eps
model2 = pd.DataFrame(data={'X2':X2,'Y2':Y2,'Eps':eps})
plt.scatter(model2['X2'],model2['Y2'])
```
